Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Corrected code to filter and select specific columns
# Sorting 'average_math' in descending order
best_math_schools = schools[schools["average_math"] >= (0.8 * 800)]
best_math_schools = best_math_schools.sort_values("average_math", ascending=False)
best_math_schools = best_math_schools[["school_name", "average_math"]]
#Views the head of best_math_schools
#best_math_schools.head()


# Add as many cells as you like...
```

```python
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"] 

top_10_schools = schools.sort_values("total_SAT", ascending = False)
top_10_schools = top_10_schools[:10]
top_10_schools = top_10_schools[["school_name", "total_SAT"]]
```

```python
# Count the number of schools in the Manhattan borough
manhattan_school_count = schools[schools['borough'] == 'Manhattan'].shape[0]
manhattan_school_count

#Average SAT scores in Manhattan
average_SAT = schools[schools["borough"] == "Manhattan"]["total_SAT"].mean()
average_SAT = round(average_SAT, 2)

#Standard deviation for SAT scores in Manhattan
std_SAT = schools[schools["borough"] == "Manhattan"]["total_SAT"].std()
std_SAT = round(std_SAT, 2)
```

```python
largest_std_dev = pd.DataFrame({
    "borough": ["Manhattan"],
    "num_schools": [manhattan_school_count],
    "average_SAT": [average_SAT],
    "std_SAT": [std_SAT]
})
largest_std_dev.set_index("borough", inplace=True)
```
